```python
# Firstly, some necessary imports

# Jupyter notebook related
%reload_ext autoreload
%autoreload 2
%matplotlib inline

# Built-in modules
import glob
import os

# Basics of Python data handling and visualization
import pickle as pkl
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
```

## Plot histogram of acquisition dates for Gillam

```python
location = 'gillam_mb_canada'

home_dir = os.environ['HOME']
data_dir = f'{home_dir}/data/northern-cities'
eopatches_dir = f'{data_dir}/{location}/eopatches_clean'

num_eopatches = len(glob.glob(f'{eopatches_dir}/eopatch-*'))

timestamp_list = []

for idx in tqdm(range(num_eopatches)):
    with open(f'{eopatches_dir}/eopatch-{idx:04d}/timestamp.pkl', 'rb') as file:
        eopatch_timestamp = pkl.load(file)
    timestamp_list.extend(eopatch_timestamp)

days = [date.timetuple().tm_yday for date in timestamp_list]
bins = [day for day in range(1, 367)]

plt.figure(figsize=(10, 5))
plt.hist(days, bins=bins);
plt.title('Acquisition dates for EOPatches of Gillam MB Canada', fontsize=14);
plt.xlabel('Days', fontsize=12);
plt.ylabel('Number of EOPatches', fontsize=12);
```
