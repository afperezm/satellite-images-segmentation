```python
# Jupyter notebook related
%reload_ext autoreload
%autoreload 2
%matplotlib inline

# Built-in modules
import datetime
import glob
import os

# Basics of Python data handling and visualization
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm

# Imports from eo-learn and sentinelhub-py
from eolearn.core import EOPatch
```

## Build Mosaic

```python
location = 'gillam_mb_canada'
# location = 'mystery_lake_mb_canada'
# location = 'yellowknife_nt_canada'
date = datetime.datetime(2020, 6, 1)

home_dir = os.environ['HOME']
data_dir = f'{home_dir}/data/northern-cities'
eopatches_dir = f'{data_dir}/{location}/eopatches_clean'

num_eopatches = len(glob.glob(f'{eopatches_dir}/eopatch-*'))

for idx in tqdm(range(num_eopatches)):

    eopatch_path = f'{eopatches_dir}/eopatch-{idx:04d}'
    eopatch = EOPatch.load(eopatch_path, lazy_loading=True)

    dates = np.array([timestamp.replace(tzinfo=None) for timestamp in eopatch.timestamp])
    closest_date_id = np.argsort(abs(date - dates))[0]

    print(f'{location}_{eopatch.timestamp[closest_date_id].strftime("%Y-%m-%d-%H-%M-%S")}_eopatch-{idx:04d}.tif')

    del eopatch
```

## Visualize EOPatches

```python
location = 'gillam_mb_canada'
date = datetime.datetime(2020, 6, 1)

a = 0.0
b = 1.0
lower_percent = 0
higher_percent = 100

home_dir = os.environ['HOME']
data_dir = f'{home_dir}/data/northern-cities'
eopatches_dir = f'{data_dir}/{location}/eopatches_clean'

fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(20, 20))

for idx in tqdm(range(15, 35)):

    eopatch_path = f'{eopatches_dir}/eopatch-{idx:04d}'
    eopatch = EOPatch.load(eopatch_path, lazy_loading=True)

    dates = np.array([timestamp.replace(tzinfo=None) for timestamp in eopatch.timestamp])
    closest_date_id = np.argsort(abs(date - dates))[0]

    ax = axes[4 - (idx - 15) % 5][(idx - 15) // 5]

    bands = eopatch.data['FEATURES'][closest_date_id][..., [2, 1, 0]]

    print(idx, eopatch.timestamp[closest_date_id], closest_date_id)

    out = np.zeros_like(bands)
    num_bands = bands.shape[2]
    for i in range(num_bands):
        band = bands[:, :, i]
        c = np.percentile(bands[:, :, i], lower_percent)
        d = np.percentile(bands[:, :, i], higher_percent)
        band = a + (band - c) * (b - a) / (d - c)
        band[band < a] = a
        band[band > b] = b
        out[:, :, i] = band

    ax.imshow(out)

    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_aspect("auto")

    del eopatch

fig.subplots_adjust(wspace=0, hspace=0)
```
